```python
import streamlit as st
import time
# --- Page Config ---
st.set_page_config(page_title="Live Time & Greeting", layout="centered")

# --- Title ---
st.title("üïí Live Time & Greeting App")

# --- Infinite Time Update ---
placeholder = st.empty()  # placeholder to update live time

while True:
    # Get current time (12-hour format with AM/PM)
    waqt = time.strftime("%d-%m-%Y | %I:%M:%S %p")
    hour = int(time.strftime("%H"))

    # Decide background color based on time
    if 6 < hour < 18:
        # Day theme
        page_bg = """
        <style>
        [data-testid="stAppViewContainer"] {
            background: linear-gradient(to bottom right, #f0f8ff, #e6f7ff);
            color: #000000;
        }
        [data-testid="stHeader"] {background: rgba(0,0,0,0);}
        </style>
        """
        greeting = "üåû Good Morning!!" if hour < 12 else "‚òÄÔ∏è Good Afternoon!!"
    elif 18 <= hour < 20:
        # Evening theme
        page_bg = """
        <style>
        [data-testid="stAppViewContainer"] {
            background: linear-gradient(to bottom right, #ffd89b, #19547b);
            color: #ffffff;
        }
        [data-testid="stHeader"] {background: rgba(0,0,0,0);}
        </style>
        """
        greeting = "üåá Good Evening!!"
    else:
        # Night theme
        page_bg = """
        <style>
        [data-testid="stAppViewContainer"] {
            background: linear-gradient(to bottom right, #0b0c10, #1f2833);
            color: #ffffff;
        }
        [data-testid="stHeader"] {background: rgba(0,0,0,0);}
        </style>
        """
        greeting = "üåô Good Night!!"

    # Apply background
    st.markdown(page_bg, unsafe_allow_html=True)

    # Update live time and greeting
    with placeholder.container():
        st.markdown(f"### üóìÔ∏è Current Time: **{waqt}**")
        st.subheader(greeting)

    # Wait for 1 second before refreshing
    time.sleep(1)
```
